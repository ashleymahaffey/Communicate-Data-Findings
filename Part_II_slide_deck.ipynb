- Insights derived from the exploration of peer-to-peer lending data.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
prosper_loan_data = pd.read_csv('/Users/ashleymahaffey/Desktop/Project Template/prosperLoanData.csv')
```

## Visualization 1: Relationship between Credit Grade and Borrower APR

## Comment:
This boxplot visualizes the relationship between Credit Grade and Borrower APR. It allows us to observe the distribution of APR across different credit grades. Higher credit grades tend to have lower median APR, indicating a negative correlation between creditworthiness and APR. The polished design includes clear labels and a title for better interpretation. 

```python
# Create a box plot to visualize the relationship between Credit Grade and Borrower APR
plt.figure(figsize=(12, 8))

# Order of CreditGrade categories for better visualization
credit_grade_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']

sb.boxplot(data=prosper_loan_data, x='CreditGrade', y='BorrowerAPR', palette='viridis', order=credit_grade_order)

# Set the title for the plot
plt.title('Relationship between Credit Grade and Borrower APR')

# Label the axes for clarity
plt.xlabel('Credit Grade')
plt.ylabel('Borrower APR (%)')

# Display the plot
plt.show()
```

## Visualization 2: Scatter plot of Loan Amount and Borrower APR

## Comment:
This scatter plot explores the relationship between Loan Amount and Borrower APR. Utilizing a log scale on the x-axis enhances the visualization, emphasizing the non-linear dynamics between these variables. The alpha parameter adds transparency for improved visibility of overlapping points. Clear labels and a title enhance interpretability.

```python
# Create a scatter plot to explore the relationship between Loan Amount, Borrower APR, and Prosper Rating
plt.figure(figsize=(12, 8))
sb.scatterplot(data=prosper_loan_data, x='LoanOriginalAmount', y='BorrowerAPR', alpha=0.3)

# Set the title for the plot
plt.title('Scatter plot of Loan Amount and Borrower APR')

# Label the axes for clarity
plt.xlabel('Loan Amount ($)')
plt.ylabel('Borrower APR (%)')

# Apply log scale to x-axis for better visibility of details
plt.xscale('log')

# Display the plot
plt.show()
```

## # Visualization 3: Distribution of Borrower APR based on Loan Status

## Comment:
This boxplot visualizes the distribution of Borrower APR based on Loan Status. It provides insights into the variation of APR across different loan statuses. The exclusion of outliers enhances focus on the main distribution. Clear labels, a title, and rotated x-axis labels improve readability and interpretation.

```python
# Create a box plot to analyze the distribution of Borrower APR based on Loan Status
plt.figure(figsize=(12, 8))

# Order of LoanStatus categories for better visualization
loan_status_order = ['Current', 'Completed', 'FinalPaymentInProgress', 'PastDue(1-15 days)', 'PastDue(15-30 days)', '...', 'Chargedoff', 'Default']

sb.boxplot(data=prosper_loan_data, x='LoanStatus', y='BorrowerAPR', palette='coolwarm', showfliers=False, order=loan_status_order)

# Set the title for the plot
plt.title('Distribution of Borrower APR based on Loan Status')

# Label the axes for clarity
plt.xlabel('Loan Status')
plt.ylabel('Borrower APR (%)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```
